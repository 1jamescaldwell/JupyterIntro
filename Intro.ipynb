a - new cell above

dd - delete cell
m - markdown
y - python
z - undo
c/x/v - copy/cut/ paste

This code creates a plot to demonstrate using numpy and Matplotlib to create a basic plot with legend, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,10,500)
y = np.cumsum(np.random.randn(500,6), 0) # the 6 here creates 6 random linespaces of y, which is why there are 6 plot lines

plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left')
```
